```python
class testClass:
    def __init__(self, name:str='', value:float=0.0):
        self.name = name
        self.value = value

    def __add__(self, other):
        if isinstance(other, testClass):
            return testClass(self.name + other.name, self.value + other.value)
        raise Exception("Can only add testClass instances")
    
    def __repr__(self):
        return f'testClass(name={self.name}, value={self.value})'

    def __str__(self):
        return f'TestClass with name: {self.name}'
```

```python
from abc import ABC, abstractmethod
import pickle

# Data Structures Examples

# List
example_list = [1, 2, 3, 4]

# Tuple
example_tuple = (1, 2, 3, 4)

# Set
example_set = {1, 2, 3, 4}

# Dictionary
example_dict = {'a': 1, 'b': 2}

# OOP 4 Pillars

# 1. Encapsulation
class Encapsulated:
    def __init__(self, value):
        self.__value = value  # private attribute

    def get_value(self):
        return self.__value

# 2. Abstraction
class AbstractBase(ABC):
    @abstractmethod
    def do_something(self):
        pass

class Concrete(AbstractBase):
    def do_something(self):
        return "Did something!"

# 3. Inheritance
class Parent:
    def greet(self):
        return "Hello from Parent"

class Child(Parent):
    def greet(self):
        return "Hello from Child"

# 4. Polymorphism
def speak(animal):
    return animal.speak()

class Dog:
    def speak(self):
        return "Woof!"

class Cat:
    def speak(self):
        return "Meow!"

# Early vs Late Binding

# Early binding: method is resolved at compile time (not typical in Python, but can be shown with direct calls)
class Early:
    def say(self):
        return "Early binding"

e = Early()
result_early = e.say()  # resolved directly

# Late binding: method is resolved at runtime (common in Python with dynamic dispatch)
def late_binding(obj):
    return obj.say()

class Late:
    def say(self):
        return "Late binding"

l = Late()
result_late = late_binding(l)  # resolved at runtime

# Functions
def add(a, b):
    return a + b

# Custom List Implementation using Dunder Methods
class CustomList:
    def __init__(self):
        self.data = []

    def __getitem__(self, idx):
        return self.data[idx]

    def __setitem__(self, idx, value):
        self.data[idx] = value

    def __len__(self):
        return len(self.data)

    def append(self, value):
        self.data.append(value)

# Custom Dictionary Implementation
class CustomDict:
    def __init__(self):
        self._keys = []
        self._values = []

    def __setitem__(self, key, value):
        if key in self._keys:
            idx = self._keys.index(key)
            self._values[idx] = value
        else:
            self._keys.append(key)
            self._values.append(value)

    def __getitem__(self, key):
        if key in self._keys:
            idx = self._keys.index(key)
            return self._values[idx]
        raise KeyError(key)

    def __contains__(self, key):
        return key in self._keys

# File Handling Basics
with open('example.txt', 'w') as f:
    f.write('Hello, file!')

with open('example.txt', 'r') as f:
    file_content = f.read()

# Serialization Basics

data = {'x': 1, 'y': 2}
with open('data.pkl', 'wb') as pf:
    pickle.dump(data, pf)

with open('data.pkl', 'rb') as pf:
    loaded_data = pickle.load(pf)
```
